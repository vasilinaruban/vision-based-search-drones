```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt") 
```

```python
# —á—Ç–æ –º–µ–Ω—è–µ–º, —Ç–∞–∫ –∏ –Ω–∞–∑—ã–≤–∞–µ–º –ø—Ä–æ–µ–∫—Ç
g_param = 'box_12'

# Train the model
results = model.train(data="data.yaml", 
                      epochs=100,
                      imgsz=640,
                      project = g_param, # –º–µ–Ω—è–µ–º
                      pretrained = True,
                      single_cls = True,
                      flipud = 0.5, # flipUpDown = 50%, fliplr = 50% - default
                      box = 12,
                     )
```

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
print('map', metrics.box.map) # map50-95
print('map50', metrics.box.map50) # map50
print('map75', metrics.box.map75)  # map75
print('precision', metrics.box.p) # precision
print('recall', metrics.box.r) # recall
```
